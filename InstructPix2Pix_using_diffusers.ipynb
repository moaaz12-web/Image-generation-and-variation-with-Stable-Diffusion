Load the `StableDiffusionInstructPix2PixPipeline` pipeline

```python
import PIL
import requests
import torch
from diffusers import StableDiffusionInstructPix2PixPipeline, EulerAncestralDiscreteScheduler
import gradio as gr

model_id = "timbrooks/instruct-pix2pix"
pipe = StableDiffusionInstructPix2PixPipeline.from_pretrained(model_id, torch_dtype=torch.float16, revision="fp16", safety_checker=None)
pipe.to("cuda")
pipe.enable_attention_slicing()
```

```python
from PIL import Image
import numpy as np
```

```python
def generate(instruction, input_image, steps):
  img1 = Image.fromarray(np.asarray(input_image))

  init_image = img1.resize((768, 512))
  if type(steps) is not int:
    try:
      steps = int(steps)
    except:
      steps = 60

  return pipe(instruction, image=init_image, num_inference_steps= steps).images[0]
```

```python
import gradio as gr
     
gr.Interface(
    
    generate,
    title = 'Image editing using InstructPIX2PIX',
    inputs=[
        gr.Textbox(label="Prompt"),
        gr.Image(elem_id = "input-image"),
        gr.Number(label = "Inference steps (higher the steps, more detailed result but it will take more time!) "),
    ],
    outputs = [
        gr.Image(elem_id="output-image")
        ], css = "#output-image, #input-image, #image-preview {border-radius: 40px !important; background-color : gray !important;} "
).launch(debug=True)
```
