```python
import torch
from diffusers import StableDiffusionPipeline

pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4", torch_dtype=torch.float16)  
```

```python
pipe = pipe.to("cuda")
```

```python

import gradio as gr
def generate(title):
  if not title or title == "" or title == " ":

    title = "A boy standing on the moon" 
  image = pipe(title).images[0]
  return image

gr.Interface(
    
    generate,
    title = 'Text to Image using Diffusers',
    inputs=[
        
        gr.Textbox(label="Prompt"),
        
    ],
    outputs = [
        gr.Image(elem_id="output-image")
        ], css = "#output-image, #input-image, #image-preview {border-radius: 40px !important; background-color : gray !important;} "
).launch(debug=True)
```
